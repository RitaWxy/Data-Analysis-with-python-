```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://beijing.anjuke.com/sale/p1'
header = {'user-agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.97 Safari/537.36'}
html = requests.get(url,headers = header)
content = html.text
```

```python
soup = BeautifulSoup(content,'lxml',from_encoding = 'utf-8')
print(soup.title)
```

```python
house_list = soup.find_all('li',class_= 'list-item')
house_names = []
house_type = []
house_space = []
house_address = []
house_price = []
house_uniprice = []

for house in house_list:
    name = house.find('div',class_ = "house-title").a.text.strip()
    house_names.append(name)
    
    type_ = house.find('div',class_ = 'details-item').span.text.strip()
    house_type.append(type_)
    
    address = house.find('span',class_ = 'comm-address').string.strip()
    house_address.append(address)
    
    price = house.find('span',class_ = 'price-det').text.strip()
    house_price.append(price)
    
    uni_price = house.find('span',class_ = 'unit-price').string.strip()
    house_uniprice.append(price)
```

```python
import pandas as pd
house_info = {'House_name          ':house_names,'House_type     ':house_type,'House_address':house_address,'House_price':house_price,'House_unitprice':house_uniprice}
house_info = pd.DataFrame(house_info)
print(house_info)
```
